```python
import pandas as pd
df = pd.read_csv('yelp_labelled.txt', names=['sentences', 'label'], sep='\t')
df.tail()
```

```python
from sklearn.model_selection import train_test_split
X = df['sentences'].values
y = df['label'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=250, oov_token='x')
tokenizer.fit_on_texts(X_train)
tokenizer.fit_on_texts(X_test)

seks_train = tokenizer.texts_to_sequences(X_train)
seks_test = tokenizer.texts_to_sequences(X_test)

pad_train = pad_sequences(seks_train, maxlen=20)
pad_test = pad_sequences(seks_test, maxlen=20)
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
                             tf.keras.layers.Embedding(250, 16, input_length=20),
                             tf.keras.layers.GlobalAveragePooling1D(),
                             tf.keras.layers.Dense(24, activation='relu'),
                             tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(pad_train, y_train, 
                 epochs=100,
                 validation_data=(pad_test, y_test), 
                 verbose=2)
```
